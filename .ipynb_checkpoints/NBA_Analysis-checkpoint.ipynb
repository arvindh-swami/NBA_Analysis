```python
# Average Career Length for players based on decade
# Calculate 
career_length = {}
for index, row in player_data.iterrows():
    print(row['year_start'])
    key = row['year_start'] - (row['year_start']%10)
    length = row['year_end'] - row['year_start'] + 1
    print(length)
    career_length[key] = career_length.setdefault(key,0) + 1
```
