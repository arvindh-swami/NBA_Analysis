```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

champions = pd.read_csv("datasets/championsdata.csv")
```

```python
champions.head()
```

```python
# How often does the team with homecourt advantage win the championship?
# Calculate number of championship wins by home team
home = 0
away = 0
year = 1980 # start year
for index, row in champions.iterrows():
    if year == row['Year']:
        if champions.loc[index]['Home'] == 1:
            home += 1
        else:
            away += 1
        year += 1
```

```python
# Bar Plot of championship wins
plt.bar(['Home', 'Away'], [home, away], width=0.5)
plt.ylabel('Number of Wins', fontsize=15)
plt.title('Championship wins of home vs away team')
plt.show()
```
